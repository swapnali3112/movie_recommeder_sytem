```python
# "[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'"
# i want this in form ------>
# ["action","adventure","fantasy","scifi"]
# we need to create helper function 
# and the list in string format so we also need to change it  with the help of ast module(literal_eval)
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l    

    
     
```

```python
dataset['genres']=dataset['genres'].apply(convert)
```

```python
dataset.head()
```

```python
dataset['keywords']=dataset['keywords'].apply(convert)
```

```python
dataset['keywords']
```

```python
# we are going to do same with cast but we want only first 3 dictionaries (it contain main characters of movies )
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i["name"])
            counter+=1
        else:
            break
        
    return l

              
```

```python
dataset["cast"]=dataset["cast"].apply(convert3)
```

```python
dataset["cast"]
```

```python
dataset.head()
```

```python
# we have done cast,crew,genres now lets do crew but its little bit difficult as compared to those 3
# we just want the data(dictinary) where job value is director 
dataset["crew"][0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break 
    return l
```

```python
dataset["crew"]=dataset["crew"].apply(fetch_director)
```

```python
# its also string so we have to convert it into list 
dataset["overview"]=dataset["overview"].apply(lambda x: x.split())

```

```python
# lets concatenate crew ,cast,keywords,genres 
# but first we need to apply transformation  for example
# "sam worthington" to "samworthington"
```

```python
dataset["genres"]=dataset["genres"].apply(lambda x: [i.replace(" ","") for i in x ])
dataset["keywords"]=dataset["keywords"].apply(lambda x: [i.replace(" ","") for i in x ])
dataset["cast"]=dataset["cast"].apply(lambda x: [i.replace(" ","") for i in x ])
dataset["crew"]=dataset["crew"].apply(lambda x: [i.replace(" ","") for i in x ])
```

```python
def recommend(movie):
    movie_index=df[df["title"] ==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key= lambda x :x[1])[1:6]
    for i in movies_list:
        print(i[0])
```
